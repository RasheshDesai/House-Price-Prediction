# Data  Visualization

```python

df_sorted = df.sort_values(by='price', ascending=False)

# Create the bar graph
plt.figure(figsize=(10, 6))
x = df_sorted['city']
y = df_sorted['price']
plt.bar(x, y)
plt.xlabel("City")
plt.ylabel("Price of House")
plt.title("City vs Price of a House (Sorted by Price)")
plt.xticks(rotation=90)  
plt.show()
```

```python
avg_price = df.groupby('city')['price'].mean()


avg_sqft_lot = df.groupby('city')['sqft_lot'].mean()


fig, ax1 = plt.subplots(figsize=(12, 6))


color = 'tab:blue'
ax1.bar(avg_price.index, avg_price, color=color, alpha=0.7, width=0.4, label='Average Price')
ax1.set_xlabel('City')
ax1.set_ylabel('Average Price', color=color)
ax1.tick_params(axis='y', labelcolor=color)


ax2 = ax1.twinx()
color = 'tab:red'
ax2.scatter(avg_sqft_lot.index, avg_sqft_lot, color=color, marker='o', label='Average sqft_lot')
ax2.set_ylabel('Average sqft_lot', color=color)
ax2.tick_params(axis='y', labelcolor=color)

for i in range(len(avg_sqft_lot) - 1):
    ax2.plot([avg_sqft_lot.index[i], avg_sqft_lot.index[i+1]], [avg_sqft_lot[i], avg_sqft_lot[i+1]], color='gray', linestyle='-')


plt.title('City vs Average Price and Average sqft_lot')
fig.legend(loc='upper left')

ax1.grid(axis='y', linestyle='--')


plt.setp(ax1.xaxis.get_majorticklabels(), rotation=90)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
x = df['sqft_lot'] + df['sqft_living']  # Total square footage of the house (house + lot)
y = df['price']
plt.scatter(x, y)
plt.xlabel("Total Square Footage of House")
plt.ylabel("Price of House")
plt.title("Square Footage of House vs Price of a House")
plt.show()
```

```python

df_sorted = df.sort_values(by='price', ascending=False)


plt.figure(figsize=(10, 6))
x = df_sorted['yr_built']
y = df_sorted['price']
plt.bar(x, y)
plt.xlabel("Year Built")
plt.ylabel("Price of House")
plt.title("Year vs Price of a House (Sorted by Price)")
plt.xticks(rotation=90)  
plt.show()
```
